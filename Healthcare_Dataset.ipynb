```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x='Insurance Provider', y='Billing Amount', data=df)
plt.title('Billing Amount by Insurance Provider')
plt.show()
```

```python
Q1 = df['Billing Amount'].quantile(0.25)
Q3 = df['Billing Amount'].quantile(0.75)

IQR = Q3 - Q1

outliers = df[(df['Billing Amount'] < Q1 - 1.5*IQR) | (df['Billing Amount'] > Q3 + 1.5*IQR)]
outliers
```

```python
df['Medical Condition'].hist(bins=12)
plt.title('Medical Condition Count')
plt.xlabel('Medical condition')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Admission Type', data=df)
plt.title('Type Count')
plt.xlabel('Admission Type')
plt.ylabel('Count')
plt.show()
```

```python
df['Test Results'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('dni Metodlar Pay')
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
df['Admission Year'] = df['Date of Admission'].dt.year
sns.countplot(x='Admission Year', data=df)
plt.title('Number of Admissions per Year')
plt.show()

```

```python
sns.catplot(x='Medical Condition', hue='Test Results', kind='count', data=df, height=5, aspect=2)
plt.xticks(rotation=45)
plt.title('Medical Condition vs Test Results')
plt.show()
```

```python
df.groupby('Gender')['Billing Amount'].mean()
```

```python
billing_stats = df.groupby('Medical Condition')['Billing Amount'].agg(['mean','median','max','min'])
billing_stats
```

```python
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days

stay_stats = df.groupby(['Medical Condition','Admission Type'])['Length of Stay'].mean().sort_values(ascending=False)
stay_stats
```

```python
# Age group yarat
bins = [0, 18, 35, 50, 65, 100]
labels = ['0-18','19-35','36-50','51-65','66+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Age Group vs Medical Condition
age_condition = df.groupby(['Age Group','Medical Condition']).size().unstack(fill_value=0)
age_condition
```

```python
doctor_stats = df.groupby('Doctor')['Billing Amount'].agg(['count','mean']).sort_values(by='count', ascending=False).reset_index()
doctor_stats
```

```python
med_test = df.groupby(['Medication','Test Results']).size().unstack(fill_value=0)
med_test_percent = med_test.div(med_test.sum(axis=1), axis=0) * 100
med_test_percent
```

```python
pivot = df.pivot_table(index='Age Group', columns='Medical Condition', values='Billing Amount', aggfunc='mean')

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title('Average Billing Amount by Age Group and Medical Condition')
plt.show()
```

```python
# Monthly average billing
df['Admission Month'] = df['Date of Admission'].dt.to_period('M')
monthly_billing = df.groupby('Admission Month')['Billing Amount'].mean()
monthly_billing.plot(figsize=(12,4), marker='o')
plt.title('Average Billing Amount per Month')
plt.show()
```
